```python

# minimax_tictactoe.py
import copy
import math
import numpy as np

X = "X"
O = "O"
EMPTY = None

user = None
ai = None

def initial_state():
    return [[EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY],
            [EMPTY, EMPTY, EMPTY]]

def player(board):
    count = 0
    for i in board:
        for j in i:
            if j:
                count += 1
    return ai if count % 2 != 0 else user

def actions(board):
    res = set()
    for i in range(3):
        for j in range(3):
            if board[i][j] == EMPTY:
                res.add((i, j))
    return res

def result(board, action):
    i, j = action
    if board[i][j] != EMPTY:
        raise Exception("Invalid action")
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board

def get_horizontal_winner(board):
    for row in board:
        if row[0] is not None and row.count(row[0]) == 3:
            return row[0]
    return None

def get_vertical_winner(board):
    for c in range(3):
        col = [board[r][c] for r in range(3)]
        if col[0] is not None and col.count(col[0]) == 3:
            return col[0]
    return None

def get_diagonal_winner(board):
    if board[0][0] is not None and board[0][0] == board[1][1] == board[2][2]:
        return board[0][0]
    if board[0][2] is not None and board[0][2] == board[1][1] == board[2][0]:
        return board[0][2]
    return None

def winner(board):
    return get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board)

def terminal(board):
    if winner(board) is not None:
        return True
    for row in board:
        for cell in row:
            if cell == EMPTY:
                return False
    return True

def utility(board):
    w = winner(board)
    if w == X:
        return 1
    elif w == O:
        return -1
    return 0

def maxValue(state):
    if terminal(state):
        return utility(state)
    v = -math.inf
    for action in actions(state):
        v = max(v, minValue(result(state, action)))
    return v

def minValue(state):
    if terminal(state):
        return utility(state)
    v = math.inf
    for action in actions(state):
        v = min(v, maxValue(result(state, action)))
    return v

def minimax(board):
    if terminal(board):
        return None
    curr_player = player(board)
    best_move = None
    if curr_player == X:
        best_val = -math.inf
        for action in actions(board):
            val = minValue(result(board, action))
            if val > best_val:
                best_val = val
                best_move = action
    else:
        best_val = math.inf
        for action in actions(board):
            val = maxValue(result(board, action))
            if val < best_val:
                best_val = val
                best_move = action
    return best_move

if __name__ == "__main__":
    board = initial_state()
    print("Choose your symbol: X or O")
    user = input().strip().upper()
    ai = O if user == X else X

    while True:
        print(np.array(board))
        if terminal(board):
            w = winner(board)
            if w:
                print("Game Over:", w, "wins")
            else:
                print("Game Over: Tie")
            break

        curr = player(board)
        if curr == user:
            print("Your turn ({}). Enter row and col 0..2".format(user))
            i = int(input("Row: "))
            j = int(input("Col: "))
            if board[i][j] == EMPTY:
                board = result(board, (i, j))
            else:
                print("Invalid move, try again.")
        else:
            move = minimax(board)
            print("AI ({}) moves: {}".format(ai, move))
            board = result(board, move)
```
